# 2 - Esse exercício faremos por etapas:
A) Verifique se temos colunas com dados faltantes e traga a % dos dados faltantes por coluna.

```python
#seu código aqui
(df.isnull().sum() / len(df))*100
```

B) Exclua nessa etapa as linhas das colunas com dados faltantes que você analisar que esse método é o ideal. Justifique o motivo de estar excluindo essas linhas.

```python
#Exclui os dados nulos das colunas Genero e Churn que tem uma porcentagem baixa desses dados.
df.dropna(subset=['Genero','Churn'], inplace=True)
(df.isnull().sum() / len(df))*100
```

C) Nessa etapa substitua pela média, mediana ou moda as linhas das colunas com valores faltantes que você analisou e entendeu que esse era o melhor método. Não se esqueça de justificar sua escolha.

```python
# Substituindo nulos da coluna Pagamento_Mensal pela mediana
mediana_pagamento = df['Pagamento_Mensal'].median()
df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(mediana_pagamento)
(df.isnull().sum() / len(df))*100
```

```python
# Considerei arriscado usar moda ou mediana na para substituir uma quantidade de dados faltantes tão alta, então resolvi criar uma coluna de nao_informado
df['PhoneService'] = df['PhoneService'].fillna('nao_informado')
(df.isnull().sum() / len(df))*100
```

# [EXTRA] 4 - Caso deseje, para manter um padrão na sua base, renomeie as colunas ou deixando em português, ou inglês.

```python
# Tradução
ajustes = {
    'customerid': 'id_cliente',
    'dependents': 'dependentes',
    'phoneservice': 'servico_telefone',
    'paymentmethod': 'metodo_pagamento'
}

df.rename(columns=ajustes, inplace=True)

print(df.columns.tolist())
```

```python
#Deixar todos os textos em minúsculo e remover espaços extras
colunas_texto = df.select_dtypes(include=['object']).columns

for col in colunas_texto:
    df[col] = df[col].str.lower().str.strip()

#Tradução de sim e não
mapeamento = {
    'yes': 'sim',
    'no': 'nao',
    'female': 'feminino',
    'male': 'masculino',
    'no internet service': 'sem servico de internet',
    'no phone service': 'sem servico de telefone'
}

df.replace(mapeamento, inplace=True)
```
